```python
# Collect info about total number of reads
f = open(os.path.join(directory,"summary_reads.csv"), "w")


hm = {}
hm_mapped = {}

directory = "/home/luisas/Desktop/cluster/data/01_bulk_RNA-Seq_lncRNAs_annotation/02_RNA-Seq_ribodepl/03_hisat/"
for subdir, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith("summary.txt"): 
            filepath = subdir + os.sep + file
            # Using readlines() 
            file1 = open(filepath, 'r') 
            Lines = file1.readlines() 
            for line in Lines: 
                if "reads" in line: 
                    reads=line.split()[0]
                if "aligned concordantly exactly 1 time" in line:
                    pe = line.split()[0]
            fileid = "_".join(file.split(".hisat")[0].split("_")[0:4])
            if fileid in hm: 
                hm[fileid] = int(hm[fileid])+int(reads)
                hm_mapped[fileid] = int(hm_mapped[fileid])+int(pe)
            else: 
                hm[fileid] = int(reads)
                hm_mapped[fileid] = int(pe)

```

```python
# Collect info about #reads in exonic regions
regions_directory = "/home/luisas/Desktop/cluster/data/01_bulk_RNA-Seq_lncRNAs_annotation/02_RNA-Seq_ribodepl/03b_rseqc/"
regions_files = open(os.path.join(regions_directory,"summary_reads.csv"), "w")


hm_assigned = {}
hm_exon = {}
hm_intron = {}

for subdir, dirs, files in os.walk(regions_directory):
    for file in files:
        if file.endswith("read_distribution.txt"): 
            filepath = subdir + os.sep + file
            # Using readlines() 
            file1 = open(filepath, 'r') 
            Lines = file1.readlines() 
            exons_line_id = ['CDS_Exons', 'UTR_Exons']
            for line in Lines: 
                if "Total Assigned Tags" in line: 
                    reads=line.split()[3]
                elif "CDS_Exons" in line:
                    exon = line.split()[2]
                elif "Intron" in line:
                    intron = line.split()[2]                    
            fileid = "_".join(file.split(".UMI")[0].split("_")[0:4])
            if fileid in hm_assigned: 
                hm_assigned[fileid] = int(hm_assigned[fileid])+int(reads)
                hm_exon[fileid] = int(hm_exon[fileid])+int(exon)
                hm_intron[fileid] = int(hm_intron[fileid])+int(intron)
            else: 
                hm_assigned[fileid] = int(reads)
                hm_exon[fileid] = int(exon)
                hm_intron[fileid] = int(intron)
```

```python
for key in hm: 
    f.write(key+","+str(hm[key])+", "+str(hm_mapped[key])+", "+str(round((hm_mapped[key]*100/hm[key]),2))+"%\n")
f.close()
```
